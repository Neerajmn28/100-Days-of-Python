# Project Description: Higher or Lower Game

Overview:
The Higher or Lower game is an interactive command-line project where players compare the follower counts of famous individuals or brands and guess who has more followers. The game tests the player's knowledge and intuition about popularity on social media platforms.

How the Game Works:
The game displays two randomly selected accounts (e.g., celebrities, brands, or public figures) along with their descriptions and countries.
Players must decide which account has more followers by typing "A" or "B".
If the guess is correct, the player earns a point, and the game continues with a new comparison.
If the guess is incorrect, the game ends, and the playerâ€™s final score is displayed.

```python
# import the libraries from data_new import data
import random
from logo import logo
from data_new import data
print(logo)
```

```python
def format_code(account):
  """Format the account data into printable format."""
  account_name = account['name']
  account_description = account['description']
  account_country = account['country']
  return (f"{account_name}, a {account_description}, from {account_country}")
```

```python
def check_answer(user_guess, a_followers, b_followers):
  """Takes the account data and returns the printable format"""
  if a_followers > b_followers:
    return user_guess == 'a'
  else:
    return user_guess == 'b'
```

```python
# Generate a random account from the game data
print(logo)
score = 0
game_should_continue = True
account_b = random.choice(data)

while game_should_continue:

  account_a = account_b
  account_b = random.choice(data)

  if account_a ==  account_b:
    account_b = random.choice(data)



  print(f'Compare A: {format_code(account_a)}')
  print( 'Vs' )
  print(f'Against B: {format_code(account_b)}')

  user_input = input('Who has more followers? Type "A" or "B": ').lower()

  print('\n' * 20)
  print(logo)



  a_follower_count = account_a['follower count']
  b_follower_count = account_b['follower count']
  is_correct = check_answer(user_input, a_follower_count, b_follower_count)
  print(is_correct)



  if is_correct == True:
    score += 1
    print(f'You are right! Current score: {score}')
  else:
    print(f'Sorry that\'s wrong!. Final score: {score}')
    game_should_continue = False
```
